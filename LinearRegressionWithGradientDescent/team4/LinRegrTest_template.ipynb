```python
# Importing libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing functions from `LinRegrGD_template.ipynb`
# %run LinRegrGD_template.ipynb
import LinRegrGD_template as LR
```

```python
# Load or generate the data and make sure is somehow showing a linear behaviour. 
# Since we loaded existing data, we do not set a random seed.
df = sns.load_dataset('car_crashes')
X = df['alcohol']
y = df['total']
sns.scatterplot(x=X,y=y) # We check that the data is linear
```

```python
# Converting X1 from series to a dataframe
X = X.to_frame()
```

```python
# Add a column of ones to X
X.insert(0, 'X0', np.ones(len(X)))
X.head()
```

```python
# Initialize the parameters and setting the hyperparameters
params = [1, 2.5]
alpha = 0.001
num_iters = 200
```

```python
# Run gradient descent
params, cost_values = LR.gradient_descent(X, y, params, alpha, num_iters)
print(params)
```

```python
# Plot the cost function
plt.plot(range(len(cost_values)), cost_values)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost per Iteration')
plt.show()
```

```python
# Plot the regression line
LR.plot_regression_line(X, y, params)
```
