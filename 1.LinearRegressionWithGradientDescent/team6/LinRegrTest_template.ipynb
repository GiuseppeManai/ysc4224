```python
# Regression with two variables

# This step should be reproducible so use a random seed
np.random.seed(0)

# Load or generate the data and make sure is somehow showing a linear behaviour. 
x1 = np.linspace(0, 10, 50)
x2 = np.random.uniform(0, 10, 50)
x2.sort()

y = 3 * x1 + 5 * x2 + np.random.normal(0, 3, len(x1))

# Add a column of ones to X
ones = np.ones(len(y))
X = np.column_stack([ones, x1, x2])

# Initialize the parameters
param = [-5 for i in range(X.shape[1])]

# Set the hyperparameters
learning_parameter = 0.001
num_iters = 100

# Run gradient descent
[param, mses] = gradient_descent(X, y, param, learning_parameter, num_iters)

# Plot the cost function
plt.figure(0)
plt.plot(np.arange(num_iters), mses)
plt.xlabel("Iteration Number")
plt.ylabel("MSE")
plt.title("MSE by Gradient Descent Iteration")

# Plot the regression line
plt.figure(1)
plot_regression_line(X, y, param)
```
